```python
import chromadb
```

```python
client = chromadb.Client()
collection = client.create_collection(name= "my_collection")
```

```python
collection.add(
    documents=[
        "This is the document about India",
        "This is the document about Chennai"
    ],
    ids=['id1', 'id2']
)
```

```python
all_docs = collection.get()
all_docs
```

```python
documents= collection.get(ids=['id1'])
documents
```

```python
results = collection.query(
    query_texts=['Query is about tamilnadu'],
    n_results=2
)

results
```

```python
collection.delete(ids=all_docs['ids'])
collection.get()
```

```python
collection.add(
    documents=[
        "This is the document about India",
        "This is the document about Tamilnadu"
    ],
    ids=['id3', 'id4'],
    metadatas=[
        {"url": "https://en.wikipedia.org/wiki/India"},
        {"url": "https://en.wikipedia.org/wiki/Tamil_Nadu"}
    ]
)
```

```python
results = collection.query(
    query_texts=['query about chennai'],
    n_results=2
)

results
```
